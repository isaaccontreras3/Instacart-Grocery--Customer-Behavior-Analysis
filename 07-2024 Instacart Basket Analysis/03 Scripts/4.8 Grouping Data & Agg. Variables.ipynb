• Finding the aggregated mean of the “order_number” column grouped by “department_id”

```python
#Using dot notation to find the mean of "order_number" by "deparment_id"
df.groupby('department_id').order_number.mean()
```

--------------------------------------------------------------------------------------------------

• Subset vs. Entire DF:

The subset shows slightly higher mean 'order_number' values for the departments compared to the entire dataset, suggesting more frequent orders or repeat customers in the subset. The subset doesn’t fully represent all departments, which could lead to biased conclusions if generalized.


----------------------------------------------------------------------------------------------------

Creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
#Creating a new DataFrame with all columns plus the new aggregated column
new_df = df.assign(max_order=df.groupby(['user_id'])['order_number'].transform(np.max))
```

```python
# Create a new DataFrame with all columns plus the new aggregated column (to avoid the warning)
#new_df = ords_prods_merge_freq.assign(max_order=ords_prods_merge_freq.groupby(['user_id'])['order_number'].transform('max'))
```

```python
new_df.shape
```

```python
new_df.head()
```

```python
 #Assigns a “loyalty” label to a user ID based on its corresponding max order value

new_df.loc[new_df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
new_df.loc[(new_df['max_order'] <= 40) & (new_df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
new_df.loc[new_df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking df
new_df.head(2)
```

--------------------------------------------------------------------------------------------------

• Identifiying if there’s a difference between the spending habits of the three types of customers

```python
#Groupping 'new_df' by 'loyalty_flag' and calculate basic statistics for 'prices'
loyalty_price_stats = new_df.groupby('loyalty_flag')['prices'].describe()

#Displaying the statistics
print(loyalty_price_stats)
```

New customers have the highest average spend per product, suggesting they may be purchasing higher-priced items.
Loyal customers spend the least on average per product, possibly due to familiarity with the product range.
The spending patterns across all customer types are similar, though outliers may influence the average prices.

--------------------------------------------------------------------------------------------------

• Targetting different types of spenders in their marketing campaigns 

```python
#Calculating the mean price of products for each user
new_df['avg_price'] = new_df.groupby('user_id')['prices'].transform('mean')

#Creating the spending flag based on the average price
new_df.loc[new_df['avg_price'] < 10, 'spending_flag'] = 'Low spender'
new_df.loc[new_df['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
new_df.head(2)
```

---------------------------------------------------------------------------------------------

• Creating an order frequency flag based on the median of the 'days_since_prior_order' variable.
This new flag will allow the Instacart team to identify and target users based on their ordering frequency

```python
# Calculating the median of 'days_since_prior_order' for each user
new_df['median_days_since_prior'] = new_df.groupby('user_id')['days_since_prior_order'].transform('median')

#Creating the order frequency flag based on the median days since the prior order
new_df.loc[new_df['median_days_since_prior'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
new_df.loc[(new_df['median_days_since_prior'] > 10) & (new_df['median_days_since_prior'] <= 20), 'order_frequency_flag'] = 'Regular customer'
new_df.loc[new_df['median_days_since_prior'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
